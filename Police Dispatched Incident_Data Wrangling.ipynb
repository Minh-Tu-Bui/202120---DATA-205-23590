```python
# Load the first dataset "Police Dispatched Incidents"
dispatch = pd.read_csv('Police_Dispatched_Incidents.csv')
```

```python
# Remove columns "Incident_ID", "Crime Reports", "Crash Reports", and "State" of the first dataset
dispatch1 = dispatch.drop(["Incident_ID", "Crime Reports", "Crash Reports","State"], axis = 1)
```

```python
dispatch1.dtypes
```

```python
dispatch1.describe(include = 'all')
```

```python
# Count the number of incidents per priority levels
dispatch1['Priority'].value_counts()
```

```python
# Information of the 4 incidents with the Priority level recorded as 5
dispatch1[dispatch1['Priority'] == 5]
```

```python
# Remove the 4 incidents with the Priority level recorded as 5. Most of the cases are "Follow up" and the 'Follow up' cases contains all for priority levels.
dispatch2 = dispatch1[dispatch1['Priority'] != 5]
```

```python
#Count missing values in the "Priority" column
dispatch2['Priority'].isnull().sum()
```

```python
#Import the "County Spending" dataset
spending = pd.read_csv('County_Spending.csv')
```

```python
#Drop "Account Code", "Vendor", "Vendor Number", "Zip", "Contract Number", 
#"PO Number", "PO Line", "Invoice Number", "Invoice Line", "Invoice Distribution Line", 
#"Payment Method", "Payment Date", "Payment Number", "Payment Status"

spending1 = spending.drop(["Account Code", "Vendor", "Vendor Number", "Zip", "Contract Number", "PO Number", "PO Line", "Invoice Number", "Invoice Line", "Invoice Distribution Line", "Payment Method", "Payment Date", "Payment Number", "Payment Status", "Invoice Date"], axis = 1)
```

```python
spending1.head()
```

```python
# Count missing values in the dataset
spending1.isnull().sum()
```

```python
# Show rows that have the Public Safety as the main service
public_safety_spending = spending1[spending1["Service"] == "Public Safety"]
```

```python
public_safety_spending
```

```python
# Total spending for Expense Categories in Public Safety service
public_safety_spending.groupby('Expense Category')['Amount'].sum()
```
